```python
def addLayer (self, ee_image_object, vis_params, name):
  map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
  folium.raster_layers.TileLayer(
    tiles = map_id_dict['tile_fetcher'].url_format,
    attr = 'Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
    name = name,
    overlay = True,
    control = True
  ).add_to(self)
```

```python
# Define a function to mosaic images by year
def mosaic_by_year (collection, starty_year, endy_year):
  yearly_images = ee.ImageCollection([])

  for year in range(starty_year, endy_year + 1):
    year_images = collection.filter(ee.Filter.calendarRange(year, year, 'year'))
    composite = year_images.mosaic()
    yearly_images = yearly_images.merge(ee.ImageCollection([composite]))
  yearly_collection = ee.ImageCollection(yearly_images)
  return yearly_collection
```
